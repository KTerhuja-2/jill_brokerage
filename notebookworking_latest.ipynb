```python
#!/usr/bin/env python3

import re
import argparse
from typing import List, Set

class TaxonomicFilter:
    def __init__(self):
        # Common taxonomic suffixes to exclude
        self.exclude_suffixes = {
            'aceae',    # Family (bacteria, fungi, plants)
            'ales',     # Order
            'idae',     # Family (animals)
            'inae',     # Subfamily
            'oideae',   # Subfamily (plants)
            'acea',     # Class
            'myceta',   # Class (fungi)
            'mycotina', # Subclass (fungi)
            'mycetes',  # Class (fungi)
            'ophyta',   # Division (plants)
            'phyceae',  # Class (algae)
            'phyta',    # Division
            'virus',    # Virus family/order
            'viridae'   # Virus family
        }

        # Higher taxonomic ranks to exclude
        self.exclude_ranks = {
            'Bacteria',
            'Archaea',
            'Fungi',
            'Virus',
            'Eukaryota',
            'Prokaryota',
            'Kingdom',
            'Phylum',
            'Class',
            'Order',
            'Family',
            'Genus'
        }

        # Compile regex patterns
        self.species_pattern = re.compile(r'^[A-Z][a-z]+ [a-z]+')  # Basic species format
        self.strain_pattern = re.compile(r'^[A-Z][a-z]+ (sp\.|strain|subsp\.)')  # Strain designations
        self.candidatus_pattern = re.compile(r'^Candidatus [A-Z][a-z]+ [a-z]+')  # Candidatus species

    def is_species_level(self, classification: str) -> bool:
        """
        Determine if a classification is at species level using multiple criteria.

        Args:
            classification (str): The taxonomic classification string

        Returns:
            bool: True if classification is at species level, False otherwise
        """
        # Remove taxid portion for checking
        name_part = classification.split('(taxid')[0].strip()

        # Check if it's a higher taxonomic rank we want to exclude
        if any(rank in name_part for rank in self.exclude_ranks):
            return False

        # Check for excluded suffixes
        if any(name_part.endswith(suffix) for suffix in self.exclude_suffixes):
            return False

        # Check if it matches any of our species patterns
        is_species = any([
            # Standard binomial nomenclature
            bool(self.species_pattern.match(name_part)),
            # Strain or species designation
            bool(self.strain_pattern.match(name_part)),
            # Candidatus species
            bool(self.candidatus_pattern.match(name_part)),
            # Additional specific cases
            'strain' in name_part.lower(),
            'isolate' in name_part.lower(),
            ' sp. ' in name_part and any(c.islower() for c in name_part),  # Species with additional info
            'serovar' in name_part.lower(),
            'var.' in name_part
        ])

        return is_species

    def filter_kraken_output(self, input_lines: List[str]) -> List[str]:
        """
        Filter Kraken2 output to show only species-level classifications.

        Args:
            input_lines (List[str]): List of Kraken2 output lines

        Returns:
            List[str]: Filtered lines containing only species-level classifications
        """
        species_entries = []

        for line in input_lines:
            if not line.strip():
                continue

            fields = line.split('\t')
            if len(fields) < 3:
                continue

            classification = fields[2]
            if self.is_species_level(classification):
                species_entries.append(line)

        return species_entries

```

```python
def main():
    # parser = argparse.ArgumentParser(description='Filter Kraken2 output for species-level classifications')
    # parser.add_argument('input_file', help='Input file containing Kraken2 output')
    # parser.add_argument('output_file', help='Output file for filtered results')
    # args = parser.parse_args()

    # Initialize filter
    tax_filter = TaxonomicFilter()

    # Process file
    try:
        with open('/content/classified/50-MN_2_output.txt', 'r') as f:
            input_lines = f.readlines()

        # Filter species entries
        species_entries = tax_filter.filter_kraken_output(input_lines)

        # Write results
        with open('/content/filtered_classified/filterd-50-MN_2_output.txt', 'w') as f:
            f.writelines(species_entries)

        print(f"Processed {len(input_lines)} entries")
        print(f"Found {len(species_entries)} species-level classifications")

    except Exception as e:
        print(f"Error processing file: {str(e)}")

if __name__ == "__main__":
    main()
```
